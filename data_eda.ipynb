```python
import pandas as pd
```

```python
data = pd.read_csv('pokemon_data_pokeapi.csv')
data.head()
```

```python
legendary_data = data[data['Legendary Status'] == 'Yes']
legendary_data
```

```python
legendary_data.describe()
```

```python
legendary_data.isna().sum()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x = legendary_data['Weight (kg)'], y= legendary_data['Height (m)'])
plt.xlabel("weight")
plt.ylabel("height")
plt.show()
```
